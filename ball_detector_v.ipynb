```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

import cv2
import numpy as np

def nothing(x):
    pass

def resize_to_fit(img, max_width=900, max_height=700):
    h, w = img.shape[:2]
    scale = min(max_width / w, max_height / h)

    if scale >= 1:
        return img  # no upscaling

    new_size = (int(w * scale), int(h * scale))
    return cv2.resize(img, new_size, interpolation=cv2.INTER_AREA)

def interactive_hsv_filter(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print("Image not found")
        return

    img = resize_to_fit(img)

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    cv2.namedWindow("Controls", cv2.WINDOW_NORMAL)

    # Create trackbars for lower and upper bounds
    cv2.createTrackbar("H_low", "Controls", 100, 179, nothing)
    cv2.createTrackbar("S_low", "Controls", 100, 255, nothing)
    cv2.createTrackbar("V_low", "Controls", 40, 255, nothing)

    cv2.createTrackbar("H_high", "Controls", 120, 179, nothing)
    cv2.createTrackbar("S_high", "Controls", 255, 255, nothing)
    cv2.createTrackbar("V_high", "Controls", 255, 255, nothing)

    while True:
        # Get current positions
        h_low = cv2.getTrackbarPos("H_low", "Controls")
        s_low = cv2.getTrackbarPos("S_low", "Controls")
        v_low = cv2.getTrackbarPos("V_low", "Controls")

        h_high = cv2.getTrackbarPos("H_high", "Controls")
        s_high = cv2.getTrackbarPos("S_high", "Controls")
        v_high = cv2.getTrackbarPos("V_high", "Controls")

        lower = np.array([h_low, s_low, v_low])
        upper = np.array([h_high, s_high, v_high])

        mask = cv2.inRange(hsv, lower, upper)
        result = cv2.bitwise_and(img, img, mask=mask)

        cv2.imshow("Original", img)
        cv2.imshow("Mask", mask)
        cv2.imshow("Filtered", result)

        key = cv2.waitKey(1) & 0xFF
        if key == 27:  # ESC to exit
            break

    cv2.destroyAllWindows()

def apply_nms(candidates, overlap_thresh=0.15):
    if not candidates: return []
    candidates = sorted(candidates, key=lambda x: x["score"], reverse=True)
    kept = []
    for cand in candidates:
        boxA = cand["box"]
        discard = False
        for k in kept:
            boxB = k["box"]
            xA, yA = max(boxA[0], boxB[0]), max(boxA[1], boxB[1])
            xB, yB = min(boxA[2], boxB[2]), min(boxA[3], boxB[3])
            interArea = max(0, xB - xA) * max(0, yB - yA)
            areaA, areaB = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1]), (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
            denom = min(areaA, areaB)
            if denom > 0 and (interArea / denom) > overlap_thresh:
                discard = True
                break
        if not discard: kept.append(cand)
    return [k["box"] for k in kept]

def detect_balls(path):
    import cv2
    import numpy as np
    import matplotlib.pyplot as plt

    img = cv2.imread(path)
    if img is None:
        return None, 0

    h, w = img.shape[:2]
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    _, sat, val = cv2.split(hsv)

    # --- STEP 1: TABLE MASK (BLUE CLOTH) ---
    lower_blue = np.array([100, 140, 40])
    upper_blue = np.array([105, 255, 255])
    table_mask = cv2.inRange(hsv, lower_blue, upper_blue)

    plt.figure(figsize=(10, 6))
    plt.imshow(table_mask, cmap='gray')
    plt.title("Table Mask")
    plt.axis('off')
    plt.show()

    # --- STEP 2: KEEP LARGEST CONTOUR (TABLE ONLY) ---
    cnts, _ = cv2.findContours(table_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not cnts:
        return img, 0

    table_cnt = max(cnts, key=cv2.contourArea)

    refined_mask = np.zeros_like(table_mask)
    cv2.drawContours(refined_mask, [table_cnt], -1, 255, -1)

    # Remove borders (important for Hough stability)
    safety_mask = cv2.erode(refined_mask, np.ones((10, 10), np.uint8))
    safety_mask = cv2.dilate(safety_mask, np.ones((10, 10), np.uint8))

    plt.figure(figsize=(10, 6))
    plt.imshow(safety_mask, cmap='gray')
    plt.title("Refined Table Mask")
    plt.axis('off')
    plt.show()

    # --- STEP 3: REMOVE CLOTH (KEEP ONLY OBJECTS INSIDE TABLE) ---
    # Invert cloth mask → non-cloth regions
    non_cloth_mask = cv2.bitwise_not(table_mask)

    # Combine: inside table AND not cloth
    objects_mask = cv2.bitwise_and(non_cloth_mask, safety_mask)

    # Apply to saturation channel (best contrast for balls)
    objects = cv2.bitwise_and(sat, sat, mask=objects_mask)
    objects = cv2.normalize(objects, None, 0, 255, cv2.NORM_MINMAX)

    plt.figure(figsize=(10, 6))
    plt.imshow(objects, cmap='gray')
    plt.title("Objects Inside Table (Cloth Removed)")
    plt.axis('off')
    plt.show()

    # --- STEP 4: CLEAN + BINARIZE ---
    blurred = cv2.GaussianBlur(objects, (9, 9), 2)
    _, binary = cv2.threshold(blurred, 30, 255, cv2.THRESH_BINARY)

    plt.figure(figsize=(10, 6))
    plt.imshow(binary, cmap='gray')
    plt.title("Binary Objects")
    plt.axis('off')
    plt.show()

    # --- STEP 5: HOUGH CIRCLES ---
    circles = cv2.HoughCircles(
        binary,
        cv2.HOUGH_GRADIENT,
        dp=1.1,
        minDist=35,
        param1=50,
        param2=10,
        minRadius=15,
        maxRadius=25
    )

    # --- STEP 6: SHOW RAW DETECTIONS ---
    debug_img = img.copy()
    count = 0

    if circles is not None:
        for (cx, cy, r) in np.round(circles[0]).astype(int):
            if 0 <= cx < w and 0 <= cy < h and safety_mask[cy, cx] == 255:
                cv2.circle(debug_img, (cx, cy), r, (0, 255, 0), 2)
                cv2.circle(debug_img, (cx, cy), 2, (0, 0, 255), 3)
                count += 1

    plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(debug_img, cv2.COLOR_BGR2RGB))
    plt.title(f"Raw Circles Detected: {count}")
    plt.axis('off')
    plt.show()

    return debug_img, count


if __name__ == "__main__":
    dev_dir = r"C:\Users\mateu\PycharmProjects\ComputerVision_1-Projeto\development_set"
    if not os.path.exists(dev_dir): dev_dir = os.getcwd()

    for filename in os.listdir(dev_dir):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            res_img, count = detect_balls(os.path.join(dev_dir, filename))
            if res_img is not None:
                plt.figure(figsize=(24, 14))
                plt.imshow(cv2.cvtColor(res_img, cv2.COLOR_BGR2RGB))
                plt.title(f"FINAL RESULT: {count} BALLS DETECTED", fontsize=50, color='red', fontweight='bold')
                plt.axis('off')
                plt.show()
```
