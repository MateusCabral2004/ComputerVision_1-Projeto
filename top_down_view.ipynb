```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- Setup & Utilities ---
def show(title, img, cmap='gray'):
    plt.figure(figsize=(10, 6))
    plt.imshow(img if len(img.shape) == 2 else cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap=cmap)
    plt.title(title)
    plt.axis('off')

def get_dist(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# --- Geometry Logic ---
def merge_lines(lines, angle_thresh=0.1, dist_thresh=100):
    merged = []
    for l in lines:
        x1, y1, x2, y2 = l[0]
        m, b = ((y2 - y1)/(x2 - x1), y1 - ((y2 - y1)/(x2 - x1))*x1) if x2 != x1 else (np.inf, x1)

        found = False
        for idx, (mr, br, x1r, y1r, x2r, y2r) in enumerate(merged):
            if (m != np.inf and mr != np.inf and abs(m - mr) < angle_thresh and abs(b - br) < dist_thresh) or \
               (m == np.inf and mr == np.inf and abs(b - br) < dist_thresh):
                xs, ys = [x1, x2, x1r, x2r], [y1, y2, y1r, y2r]
                merged[idx] = (mr, br, min(xs), min(ys), max(xs), max(ys))
                found = True
                break
        if not found:
            merged.append((m, b, x1, y1, x2, y2))
    return merged

def order_points(pts):
    pts.sort(key=lambda c: c[1])
    # Identify Top-Left vs Top-Right
    tl = pts[0] if get_dist(pts[0], [0, pts[0][1]]) < get_dist(pts[1], [0, pts[1][1]]) else pts[1]
    tr = pts[1] if np.array_equal(tl, pts[0]) else pts[0]
    # Identify Bottom-Left vs Bottom-Right
    bl = pts[2] if get_dist(pts[2], [0, pts[2][1]]) < get_dist(pts[3], [0, pts[3][1]]) else pts[3]
    br = pts[3] if np.array_equal(bl, pts[2]) else pts[2]
    return np.array([tl, tr, br, bl], dtype="float32")

def get_intersection(p1, p2, p3, p4):
    def line_params(pa, pb):
        m = (pb[1] - pa[1]) / (pb[0] - pa[0])
        return m, pa[1] - m * pa[0]
    m1, b1 = line_params(p1, p2)
    m2, b2 = line_params(p3, p4)
    x = (b2 - b1) / (m1 - m2)
    return int(x), int(m1 * x + b1)

# --- Main Processing ---
img = cv2.imread(os.path.join("../images", "c.jpg"))
h, w = img.shape[:2]

show("Original Image", img)

# 1. Masking & Edge Detection
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, np.array([100, 130, 40]), np.array([130, 255, 255]))

contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
blob_mask = np.zeros_like(mask)
cv2.drawContours(blob_mask, [max(contours, key=cv2.contourArea)], -1, 255, -1)

edges = cv2.Canny(blob_mask, 100, 200)
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 60, minLineLength=80, maxLineGap=200)

if lines is not None:
    # 2. Find Corners
    merged = merge_lines(lines)
    all_intersections = []
    for i in range(len(merged)):
        for j in range(i+1, len(merged)):
            m1, b1, _, _, _, _ = merged[i]
            m2, b2, _, _, _, _ = merged[j]
            if m1 == m2: continue

            x = (b2 - b1)/(m1 - m2) if m1 != np.inf and m2 != np.inf else (b1 if m1 == np.inf else b2)
            y = m1*x + b1 if m1 != np.inf else m2*x + b2

            if 0 <= x < w*0.9 and 0 <= y < h*0.9:
                all_intersections.append((int(x), int(y)))

    # Filter close points
    corners = []
    for p in all_intersections:
        if not any(np.linalg.norm(np.array(p) - np.array(c)) < 20 for c in corners):
            corners.append(p)

    # 3. Scaling & Perspective
    pts = order_points(corners)
    cx, cy = get_intersection(pts[0], pts[2], pts[1], pts[3]) # tl-br and tr-bl

    scaled = []
    max_d = get_dist((0, 0), (w//2, h))
    for x, y in pts:
        factor = 1 + 0.2 * (1 - get_dist((x, y), (w//2, h)) / max_d)
        scaled.append((int(cx + (x - cx) * factor), int(cy + (y - cy) * factor)))

    # Transform
    src = np.array([scaled[1], scaled[2], scaled[3], scaled[0]], dtype="float32") # tr, br, bl, tl
    dst = np.array([[0, 0], [1600, 0], [1600, 800], [0, 800]], dtype="float32")

    warped = cv2.warpPerspective(img, cv2.getPerspectiveTransform(src, dst), (1600, 800))
    show("Warped Result", warped)
else:
    print("No lines detected.")
```
